```python
import cv2

def convertVideo(videoFile):
    cap = cv2.VideoCapture(videoFile)
    if cap.isOpened() == False:
        print("cannot play file: ", videoFile)
        return
    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
 
    # Define the codec and create VideoWriter object.The output is stored in 'outpy.avi' file.
    out = cv2.VideoWriter(videoFile + 'outpy.mp4',cv2.VideoWriter_fourcc(*'MP4V'), 20, (frame_width,frame_height))
 
    while (cap.isOpened()):
        ret, frame = cap.read()
        if ret == True:
        # Display the resulting frame
            crop_img = frame#[(int(frame_width/2)):, :]
            # Convert to grayscale

            gray = cv2.cvtColor(crop_img, cv2.COLOR_BGR2GRAY)
            #gray = create_sepia(gray)
            # Gaussian blur
            blur = cv2.GaussianBlur(gray,(5,5),0)
            # Color thresholding

            ret,thresh = cv2.threshold(blur,120,255,cv2.THRESH_BINARY_INV)
            # Find the contours of the frame
            contours,hierarchy = cv2.findContours(thresh.copy(), 1, cv2.CHAIN_APPROX_NONE)
            # Find the biggest contour (if detected)
            if len(contours) > 0:
                c = max(contours, key=cv2.contourArea)
                M = cv2.moments(c)
                cx = int(M['m10']/M['m00'])
                cy = int(M['m01']/M['m00'])

                cv2.line(crop_img,(cx,0),(cx,720),(255,0,0),1)
                cv2.line(crop_img,(0,cy),(1280,cy),(255,0,0),1)
                cv2.drawContours(crop_img, contours, -1, (0,255,0), 1)
                if cx >= 120:
                    print("Turn Right!")
                if cx < 120 and cx > 50:
                    print("On Track!")
                if cx <= 50:
                    print("Turn Left")
            else:
                print("I don't see the line")
            cv2.imshow('Frame',crop_img)
            #crop_img = cv2.cvtColor(crop_img, cv2.COLOR_GRAY2BGR)
            out.write(crop_img)
        # Press Q on keyboard to  exit
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break
            # Break the loop
        else: 
            break
    # When everything done, release the video capture object
    cap.release()
    out.release()
    # Closes all the frames
    cv2.destroyAllWindows()
    print("Done converting")
```
